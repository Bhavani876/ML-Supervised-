# Feature Scaling

```python
def feature_scaling(scale):
    new_df=df.select_dtypes(exclude=['object','category'])
    if scale=='minmax':
        scaler=preprocessing.MinMaxScaler()
    else: 
        scaler=preprocessing.standardscaler()
    df_scale=pd.DataFrame(scaler.fit_transform(new_df),columns=new_df.columns)
    return df_scale

final_df=feature_scaling('minmax')
final_df.head()
```

# Visualize for Numerical data

```python
numerics=['int64','int32']
new_income_df=df.select_dtypes(include=numerics)
numeric_cols=new_income_df.columns
num_cols=4
bins=40
num_rows=int(len(numeric_cols)/num_cols)+1
fig, axs=plt.subplots(num_rows,num_cols,tight_layout=True,figsize=(20,10))
for col, ax in zip(numeric_cols,axs.flatten()[:len(numeric_cols)]):
    ax.hist(new_income_df[col],bins=bins,density=True)
    ax.set_title(col);
```
