```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.datasets import make_regression
```

```python
import sklearn.datasets
```

```python
x,y=make_regression(n_samples=100,n_features=1,n_targets=1, bias=4, effective_rank=None, tail_strength=0.5, noise=0.0, shuffle=True, coef=False, random_state=100)
```

```python
x
```

```python
y
```

# Visualize the toy datasets as a scatter plot.

```python
plt.figure(figsize=(10,4))
x.reshape(-1,1)
y.reshape(-1,1)
plt.scatter(x,y);
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=130)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
LR.score(x_train,y_train)
```

```python
print('Intercept',LR.intercept_)
print('Slope',LR.coef_)
```

```python
predict_y=LR.predict(x_test)
predict_y
```

# Fit a linear regression model to the dataset (line plot) and visualize it with the data.

```python
plt.figure(figsize=(10,4))
x.reshape(-1,1)
y.reshape(-1,1)
plt.scatter(x,y)
plt.plot(x_test.squeeze(),predict_y.squeeze(),color='r')
```

```python
X.shape
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=.3,random_state=100)
```

```python
print('Intercept',LR.intercept_)
print('Slope',LR.coef_)
```

```python
predict_y=LR.predict(X_test)
predict_y
```
